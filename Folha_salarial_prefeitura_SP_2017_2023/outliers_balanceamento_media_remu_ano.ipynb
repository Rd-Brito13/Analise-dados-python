```python
import pandas as pd
# Configurando o pandas para exibir todos os valores que sejam float com (, . e duas casas decimais)
pd.options.display.float_format = "{:,.2f}".format
#criando o dataframe
receita_final = pd.read_pickle('receita_final.pkl')
display(receita_final)
```

```python
#converter jornada para float e remover a STRING
receita_final['Jornada'] = receita_final['Jornada'].str.extract('(\d+)').astype(float)
#converter remunreação Bruta para float64
receita_final['Remuneração Bruta'] = pd.to_numeric(receita_final['Remuneração Bruta'], errors = 'coerce')
#exibir informações das colunas do dataframe
receita_final.info()
```

```python
#exibir dataframe
display(receita_final)
```

```python
#retornando a media das remunerações brutas e retornando a media das jornadas de trabalaho
media_salarial = receita_final['Remuneração Bruta'].mean()
media_horas = receita_final['Jornada'].mean()
print('a:{:,.2f}, b:{:,.2f}'.format(media_salarial, media_horas))
```

```python
#criando o dataframe a partir dos filtro no dataframe antgio
outliers = receita_final[(receita_final['Remuneração Bruta'] < media_salarial) & (receita_final['Jornada'] > media_horas)].reset_index(drop = True)
balanceamento = receita_final[(receita_final['Remuneração Bruta'] > media_salarial) & (receita_final['Jornada'] < media_horas)].reset_index(drop = True)
```

```python
#exibindo os outliers entre os anos de 2017 até 2023
display(outliers)
outliers.to_pickle('outliers.pkl')
#exibindo balanceamento entre so anos 2017 até 2023
display(balanceamento)
balanceamento.to_pickle('balanceamento.pkl')
```

```python
#convertendo em datefime
receita_final['Data'] = pd.to_datetime(receita_final['Data'])
#extraindo o ano
receita_final['Ano'] = receita_final['Data'].dt.year
#agrupando e retornando a media
media_remuneracao_ano = receita_final.groupby('Ano')['Remuneração Bruta'].mean().reset_index()
#exibindo
display(media_remuneracao_ano)
```

```python
#exportando o arquivo
media_remuneracao_ano.to_pickle("media_ano_remuneracao.pkl")
```
