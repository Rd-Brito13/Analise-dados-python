## Gr√°fico de linhas

```python
import matplotlib.pyplot as plt
meses = ["Janeiro","Fevereiro", "Mar√ßo", "Abril","Maio","junho","julho","Agosto","Setembro","Outubro","Novembro","Dezembro"]
vendas = [85,92,108,215,351,356,1786,15478,13695,25478,74854,96853]
#Listas que ser√£o utilizadas como base para a cria√ß√£o do gr√°fico
plt.figure(figsize=(12,5))
#fun√ß√£o para difinir o tamanho do eixo x e y do gr√°fico.
grafico_linhas = plt.plot(meses, vendas, marker="o")
#definindo o tipo do gr√°fico, como tipo linha, e dando uma marca√ß√£o em bolinhas para cada t√≥pico apresentado
plt.xlabel("m√™ses")
#definindo o rotulo da coluna
plt.ylabel("vendas")
#definindo o rotulo da linhas
plt.show()
```

## Gr√°fico de barras


```python
cidades = ["S√£o Paulo", "Rio de Janeiro", "Bahia"]
temperatura = [25,33,55]
plt.figure(figsize=(10,5))
grafico_barras = plt.bar(cidades, temperatura, color="yellow")
#definindo a variavel que ir√° abrir o gr√°fico
plt.bar_label(grafico_barras, labels=temperatura, padding=3, fontsize=10, fontweight="bold")
#formatando as defini√ß√µes internas do gr√°fico, rotulos das colunas, tipo de font, negrito, separa√ß√£o dos rotulos e das barras
plt.ylabel("Temperaturas")
#definindo o rotulo das linha
plt.xlabel("Cidades")
#definindo o rotulo das colunas
plt.show()
```

## Gr√°fico pizza ou pie


```python
categorias = ["Eletr√¥nicos", "Roupas", "Alimentos", "Livros"]
quantidade_percentual = [45,30,15,10] #valores em porcentagem
#Dados √† ser utilizado no gr√°fico
destaques = [0,0.1,0.1,0.2] 
#lista utilizada com valores pr√©-selecionados utilizados para fazer a separa√ß√£o das fatias atrav√©s do paamentro "explode"
cores = plt.get_cmap("Pastel2").colors 
#Fun√ß√£o utilizada para capturar uma palheta de cores entra varias disponiveis e passando para uma variavel est√° connfigura√ß√£o
plt.title("Ditribui√ß√£o percentual dos produtos em estoque")
#fun√ß√£o para estabelecer o titulo do gr√°fico
grafico_pizza = plt.pie(quantidade_percentual, labels=categorias,autopct="%.1f", explode=destaques, colors = cores)
#definindo o tipo do gr√°fico e configurando ele.
plt.show()
#exibindo o gr√°fico
```

## Gr√°fico barras empilhadas

```python
from cycler import cycler
anos = [2018,2019,2020,2021]
vendas_online = [5,10,15,20]
vendas_loja = [10,8,6,4]
#dados a serem utilizados no gr√°fico
cores = cycler("color", plt.get_cmap("Pastel1").colors)
plt.rc("axes",prop_cycle=cores)
#variavel que vai abrir o mapa de cores nomeado como "pastel1", a variavel ir√° se tornar uma tupla com diversos valores rgb onde cada item dessa tupla, 
#ser√° uma cor, e foi passada ela em parametros, com a rele√ß√£o dos indices para retonar a cor desejada.
#com este esquema criado utilizando a biblioteca cyler e a fun√ß√£o plt.rc, foi criado um loop infinito onde, sempre ir√° usar as cores da paleta definida na ordem.
#nunca repetindo elas e sem a necessidade de definir a cor na linha criadora do gr√°fico
grafico_barras_empilhadas = plt.bar(anos,vendas_online, label="Online" )
grafico_barras_empilhadas = plt.bar(anos,vendas_loja,bottom=vendas_online ,label="F√≠sica")
#definindo a variavel que ir√° abrir o gr√°fico e o tipo do grafico, o parametro "bottom", √© essencial para ser feito 
#o grafico empilhado, pois ser√° definido que o dado selecionado esteja em baixo e por padr√£o, o outro por cima.
grafico_barras_empilhadas = plt.xticks(ticks=anos)
#a fun√ß√£o "plt.xticks", foi utilizada para fazer a formata√ß√£o do do eixo x com os valores dos anos.
grafico_barras_empilhadas = plt.ylabel("Vendas")
#rotulo de para eixo y
grafico_barras_empilhadas = plt.xlabel("Anos")
#rotulo para eixo x
grafico_barras_empilhadas = plt.title("Vendas F√≠sica vs Online")
#titulo do gr√°fico
plt.legend()
#fun√ß√£o utilizada para representar a legende de cada label passada no inicio, na cria√ß√£o do grafico para simbolizar a defini√ß√£o de 
#cada dado representado

plt.show(grafico_barras_empilhadas)
#exibindo o grafico
```
