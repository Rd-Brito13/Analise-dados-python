## Gráfico de linhas

```python
import matplotlib.pyplot as plt
meses = ["Janeiro","Fevereiro", "Março", "Abril","Maio","junho","julho","Agosto","Setembro","Outubro","Novembro","Dezembro"]
vendas = [85,92,108,215,351,356,1786,15478,13695,25478,74854,96853]
#Listas que serão utilizadas como base para a criação do gráfico
plt.figure(figsize=(12,5))
#função para difinir o tamanho do eixo x e y do gráfico.
grafico_linhas = plt.plot(meses, vendas, marker="o")
#definindo o tipo do gráfico, como tipo linha, e dando uma marcação em bolinhas para cada tópico apresentado
plt.xlabel("mêses")
#definindo o rotulo da coluna
plt.ylabel("vendas")
#definindo o rotulo da linhas
plt.show()
```

## Gráfico de barras


```python
cidades = ["São Paulo", "Rio de Janeiro", "Bahia"]
temperatura = [25,33,55]
plt.figure(figsize=(10,5))
grafico_barras = plt.bar(cidades, temperatura, color="yellow")
#definindo a variavel que irá abrir o gráfico
plt.bar_label(grafico_barras, labels=temperatura, padding=3, fontsize=10, fontweight="bold")
#formatando as definições internas do gráfico, rotulos das colunas, tipo de font, negrito, separação dos rotulos e das barras
plt.ylabel("Temperaturas")
#definindo o rotulo das linha
plt.xlabel("Cidades")
#definindo o rotulo das colunas
plt.show()
```

## Gráfico pizza ou pie


```python
categorias = ["Eletrônicos", "Roupas", "Alimentos", "Livros"]
quantidade_percentual = [45,30,15,10] #valores em porcentagem
#Dados à ser utilizado no gráfico
destaques = [0,0.1,0.1,0.2] 
#lista utilizada com valores pré-selecionados utilizados para fazer a separação das fatias através do paamentro "explode"
cores = plt.get_cmap("Pastel2").colors 
#Função utilizada para capturar uma palheta de cores entra varias disponiveis e passando para uma variavel está connfiguração
plt.title("Ditribuição percentual dos produtos em estoque")
#função para estabelecer o titulo do gráfico
grafico_pizza = plt.pie(quantidade_percentual, labels=categorias,autopct="%.1f", explode=destaques, colors = cores)
#definindo o tipo do gráfico e configurando ele.
plt.show()
#exibindo o gráfico
```

## Gráfico barras empilhadas

```python
from cycler import cycler
anos = [2018,2019,2020,2021]
vendas_online = [5,10,15,20]
vendas_loja = [10,8,6,4]
#dados a serem utilizados no gráfico
cores = cycler("color", plt.get_cmap("Pastel1").colors)
plt.rc("axes",prop_cycle=cores)
#variavel que vai abrir o mapa de cores nomeado como "pastel1", a variavel irá se tornar uma tupla com diversos valores rgb onde cada item dessa tupla, 
#será uma cor, e foi passada ela em parametros, com a releção dos indices para retonar a cor desejada.
#com este esquema criado utilizando a biblioteca cyler e a função plt.rc, foi criado um loop infinito onde, sempre irá usar as cores da paleta definida na ordem.
#nunca repetindo elas e sem a necessidade de definir a cor na linha criadora do gráfico
grafico_barras_empilhadas = plt.bar(anos,vendas_online, label="Online" )
grafico_barras_empilhadas = plt.bar(anos,vendas_loja,bottom=vendas_online ,label="Física")
#definindo a variavel que irá abrir o gráfico e o tipo do grafico, o parametro "bottom", é essencial para ser feito 
#o grafico empilhado, pois será definido que o dado selecionado esteja em baixo e por padrão, o outro por cima.
grafico_barras_empilhadas = plt.xticks(ticks=anos)
#a função "plt.xticks", foi utilizada para fazer a formatação do do eixo x com os valores dos anos.
grafico_barras_empilhadas = plt.ylabel("Vendas")
#rotulo de para eixo y
grafico_barras_empilhadas = plt.xlabel("Anos")
#rotulo para eixo x
grafico_barras_empilhadas = plt.title("Vendas Física vs Online")
#titulo do gráfico
plt.legend()
#função utilizada para representar a legende de cada label passada no inicio, na criação do grafico para simbolizar a definição de 
#cada dado representado

plt.show(grafico_barras_empilhadas)
#exibindo o grafico
```
